```python
# Data Manipulation
import numpy as np
import pandas as pd 

# Visualization 
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight') # change the style of the plot
sns.set(rc = {'figure.figsize':(10,5)})
```

```python
data = pd.read_csv('dataset.csv')
data.index = pd.to_datetime(data.index)
data.head()
```

```python
sns.lineplot(data=data[['SP500 Close']], color="blue", linewidth = 2.5)
plt.title('S&P 500 index', fontsize=18)
plt.xlabel('Time (month)')
plt.ylabel('S&P 500 index level')
plt.legend(loc=2, prop={'size': 15})
plt.annotate('Dot Com Bubble', xy=('1999-10-01', 1500), xytext=('1995-01-01', 2000),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Financial Crisis', xy=('2009-01-01', 700), xytext=('2012-01-01', 500),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Covid19', xy=('2020-03-01', 2700), xytext=('2020-08-01', 1500),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.tight_layout()
plt.savefig('Graph/S&P500 index.png')
```

```python
macro = data[['CPIAUCSL_PC1', 'UNRATE', 'FEDFUNDS', 'T10Y2Y']]
```

```python
sns.lineplot(data=macro, palette="icefire", linewidth = 1.5)
plt.title('Macroeconomic Indicators', fontsize=18)
plt.xlabel('Time (month)')
plt.ylabel('Percentage or Change Over Time')
plt.legend(prop={'size': 12})
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Graph/Macroeconomic Indicators.png')
```
