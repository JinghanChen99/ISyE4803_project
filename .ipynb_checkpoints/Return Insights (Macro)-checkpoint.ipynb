```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight') # change the style of the plot
```

### Market Returns Insights
This notebook investigates the impact of macroeconomic conditions on the performance of the S&P 500 index. We will convert the continuous variable into several different categories and compute the lagged return under each category.

```python
data = pd.read_csv('dataset.csv')
data.index = pd.to_datetime(data.index)
data.head()
```

```python
annual = data.resample('YS').mean()
annual['SP500 annual return'] = annual['SP500 Close'].pct_change()
# annual.head()
```

```python
avg_annual_return = round(annual['SP500 annual return'].mean(), 4) * 100
print(f"The average annual return since 1976 is {avg_annual_return}%")
sns.histplot(data=annual, x='SP500 annual return', bins=20)
```

```python
month = data.resample('MS').mean()
month['SP500 monthly return'] = month['SP500 Close'].pct_change()
# month.head()
```

```python
avg_monthly_return = round(month['SP500 monthly return'].mean(), 4) * 100
print(f"The average monthly return since 1976 is {avg_monthly_return}%")
sns.histplot(data=month, x='SP500 monthly return', bins=20)
```

```python
# function that's used to annualize a monthly return 
def annualize(monthly):
    return round((1 + monthly) ** 12 - 1, 4)
```

### Inflation 
Consumer Price Index (CPI) all items to different levels: low, mid, high. These levels corresponds to the different level of inflation in an economy: creeping, walking, galloping.

```python
sns.histplot(data=annual, x="CPIAUCSL_PC1")
```

```python
annual['CPI level'] = np.where(annual['CPIAUCSL_PC1'] < 0, 'deflation (<0)',
                            np.where(annual['CPIAUCSL_PC1'] <= 2, 'healthy (<=2)', 
                               np.where(annual['CPIAUCSL_PC1'] <= 3, 'low (<=3)', 
                                       np.where(annual['CPIAUCSL_PC1'] < 10, 'mid (<10)', 'high(>10)'))))
annual_CPI_return = annual.groupby('CPI level').mean()['SP500 annual return'].to_frame()
reorderlist = ['deflation (<0)', 'healthy (<=2)', 'low (<=3)', 'mid (<10)', 'high(>10)']
annual_CPI_return = annual_CPI_return.reindex(reorderlist)
sns.barplot(data=annual_CPI_return, x=annual_CPI_return.index, y='SP500 annual return', palette="Blues_d")
plt.axhline(round(annual['SP500 annual return'].mean(), 4), color = 'red', alpha=0.4, linestyle='--', linewidth = 2)
plt.xticks(rotation=45)
plt.title('S&P 500 Annual Return vs CPI Level')
plt.tight_layout()
plt.savefig('Graph/S&P500 annual return vs CPI.png')
```

We can see from the CPI versus avg annual return plot, S&P 500 index has: 
- The worst performance in a deflationary environment. 
- Has a better than average performance in the economic environment where CPI is at healthy or low level. 
- Underperforms in a mid level inflationary environment ($ 3 \le CPI <10$)
- Just about average performance in a high inflationary environment

```python
month['CPI level'] = np.where(month['CPIAUCSL_PC1'] < 0, 'deflation (<0)',
                            np.where(month['CPIAUCSL_PC1'] <= 2, 'healthy (<=2)', 
                               np.where(month['CPIAUCSL_PC1'] <= 3, 'low (<=3)', 
                                       np.where(month['CPIAUCSL_PC1'] < 10, 'mid (<10)', 'high(>10)'))))
monthly_CPI_return = month.groupby('CPI level').mean()['SP500 monthly return'].to_frame()
reorderlist = ['deflation (<0)', 'healthy (<=2)', 'low (<=3)', 'mid (<10)', 'high(>10)']
monthly_CPI_return = monthly_CPI_return.reindex(reorderlist)
# monthly_CPI_return
```

```python
def delay_return_CPI(lag = 1):
    a = []
    b = []
    c = []
    d = []
    e = []
    for index, row in month.iterrows():
        # print(index, row)
        if row['CPI level'] == 'deflation (<0)':
            a.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['CPI level'] == 'healthy (<=2)':
            b.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['CPI level'] == 'low (<=3)':
            c.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['CPI level'] == 'mid (<10)':
            d.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['CPI level'] == 'high(>10)':
            e.append(month['SP500 monthly return'].shift(-lag).loc[index])
    a = [x for x in a if str(x) != 'nan']
    b = [x for x in b if str(x) != 'nan']
    c = [x for x in c if str(x) != 'nan']
    d = [x for x in d if str(x) != 'nan']
    e = [x for x in e if str(x) != 'nan']
    return round(np.mean(a), 4), round(np.mean(b), 4), round(np.mean(c), 4), round(np.mean(d), 4), round(np.mean(e), 4)
```

```python
D = {}
He = {}
L = {}
M = {}
Hi = {}
for i in range(-12, 13):
    d, he, l, m, hi = delay_return_CPI(i)
    D[i] = annualize(d)
    He[i] = annualize(he)
    L[i] = annualize(l)
    M[i] = annualize(m)
    Hi[i] = annualize(hi)
```

```python
D = pd.DataFrame.from_dict(D, orient='index', columns=['deflation (<0)'])
He = pd.DataFrame.from_dict(He, orient='index', columns=['healthy (<=2)'])
L = pd.DataFrame.from_dict(L, orient='index', columns=['low (<=3)'])
M = pd.DataFrame.from_dict(M, orient='index', columns=['mid (<10)'])
Hi = pd.DataFrame.from_dict(Hi, orient='index', columns=['high(>10)'])
df = pd.concat([D, He, L, M, Hi], axis = 1)
sns.lineplot(data=df, palette="icefire", linewidth = 1)
plt.title('S&P 500 Monthly Lagged Return')
plt.axhline(round(annual['SP500 annual return'].mean(), 4), color = 'red', alpha=0.4, linestyle='--', linewidth = 2)
sns.set(rc={'figure.figsize':(8,5)})
plt.xlabel("Lags (month)")
plt.ylabel("Annualized Return")
plt.legend(loc=4, prop={'size': 15})
plt.tight_layout()
plt.savefig('Graph/S&P500 month lagged return vs CPI.png')
```

On the more granular monthly level, we can track the average lagged return before and after each CPI level. 
- Deflationary environment has the most volitile return. It tends to have a bear market follow up with high returns and then dips again to the downside.
- A high inflationary enviroment tends to outperform before going below the average S&P 500 annual return after 5 months.
- A mid level inflationary enviroment underperform overall.
- A low level and healthy inflationary enviroment tend to outperform.

```python
month['unemployment level'] = np.where(month['UNRATE'] < 6, 'low',
                            np.where(month['UNRATE'] < 8, 'mid', 'high'))
monthly_UNRATE_return = month.groupby('unemployment level').mean()['SP500 monthly return'].to_frame()
reorderlist = ['low', 'mid', 'high']
monthly_UNRATE_return = monthly_UNRATE_return.reindex(reorderlist)
# monthly_UNRATE_return
```

```python
def delay_return_UNRATE(lag = 1):
    a = []
    b = []
    c = []
    for index, row in month.iterrows():
        # print(index, row)
        if row['unemployment level'] == 'low':
            a.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['unemployment level'] == 'mid':
            b.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['unemployment level'] == 'high':
            c.append(month['SP500 monthly return'].shift(-lag).loc[index])
    a = [x for x in a if str(x) != 'nan']
    b = [x for x in b if str(x) != 'nan']
    c = [x for x in c if str(x) != 'nan']
    return round(np.mean(a), 4), round(np.mean(b), 4), round(np.mean(c), 4)
```

```python
L = {}
M = {}
H = {}
for i in range(-12, 13):
    l, m, h = delay_return_UNRATE(i)
    L[i] = annualize(l)
    M[i] = annualize(m)
    H[i] = annualize(h)
```

```python
L = pd.DataFrame.from_dict(L, orient='index', columns=['low'])
M = pd.DataFrame.from_dict(M, orient='index', columns=['mid'])
H = pd.DataFrame.from_dict(H, orient='index', columns=['high'])
df = pd.concat([L, M, H], axis = 1)
sns.lineplot(data=df, palette="icefire")
plt.title('S&P500 Monthly Lagged Return vs Unemployment Level')
plt.axhline(round(annual['SP500 annual return'].mean(), 4), color = 'red', alpha=0.4, linestyle='--', linewidth=2)
sns.set(rc={'figure.figsize':(8,5)})
plt.xlabel("Lags (month)")
plt.ylabel("Annualized Return")
plt.tight_layout()
plt.savefig('Graph/S&P500 Monthly Lagged Return vs Unemployment.png')
```

On the monthly level, we can track the average lagged return before and after each unemployment level. 
- High unemployment environment tend to coexist with the most volitile market 
- A mid level unemployment tend to perform around the average market return 
- A low level unemployment tend to outperform the market before underperforming for a significant long period of time

### FED FUNDS Rate

```python
sns.histplot(data=annual, x="FEDFUNDS", bins=10)
```

```python
annual['FED Funds level'] = np.where(annual['FEDFUNDS'] < 0.5, '0 ~ 0.5',
                                np.where(annual['FEDFUNDS'] < 1, '0.5 ~ 1', 
                                     np.where(annual['FEDFUNDS'] < 1.5, '1 ~ 1.5', 
                                              np.where(annual['FEDFUNDS'] < 2, '1.5 ~ 2',
                                                      np.where(annual['FEDFUNDS'] < 3, '2 ~ 3',
                                                              np.where(annual['FEDFUNDS'] < 6, '3 ~ 6',
                                                                      np.where(annual['FEDFUNDS'] < 12, '6 ~ 12', '> 12')))))))
FEDFUNDS_return = annual.groupby('FED Funds level').mean()['SP500 annual return'].to_frame()
FEDFUNDS_return
```

```python
reorderlist = ['0 ~ 0.5', '0.5 ~ 1', '1 ~ 1.5', '1.5 ~ 2', '2 ~ 3', '3 ~ 6','6 ~ 12', '> 12']
FEDFUNDS_return = FEDFUNDS_return.reindex(reorderlist)
FEDFUNDS_return
```

```python
month['FED Funds Change'] = month['FEDFUNDS'].diff()
month[['FED Funds Change']].hist(bins=50)
```

```python
month['FED Funds Change level'] = np.where(month['FED Funds Change'] < -0.5, 'large drop',
                                    np.where(month['FED Funds Change'] < 0, 'drop', 
                                         np.where(month['FED Funds Change'] < 0.5, 'raise', 'large raise')))
month['FED Funds Change level'].value_counts()
```

```python
def delay_return_FEDFUNDS(lag = 1):
    a = []
    b = []
    c = []
    d = []
    for index, row in month.iterrows():
        # print(index, row)
        if row['FED Funds Change level'] == 'raise':
            a.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['FED Funds Change level'] == 'large raise':
            b.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['FED Funds Change level'] == 'drop':
            c.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['FED Funds Change level'] == 'large drop':
            d.append(month['SP500 monthly return'].shift(-lag).loc[index])
    a = [x for x in a if str(x) != 'nan']
    b = [x for x in b if str(x) != 'nan']
    c = [x for x in c if str(x) != 'nan']
    d = [x for x in d if str(x) != 'nan']
    return round(np.mean(a), 4), round(np.mean(b), 4), round(np.mean(c), 4), round(np.mean(d), 4)
```

```python
R = {}
LR = {}
D = {}
LD = {}
for i in range(-12, 13):
    r, lr, d, ld = delay_return_FEDFUNDS(i)
    # print(r, lr, d, ld)
    R[i] = annualize(r)
    LR[i] = annualize(lr)
    D[i] = annualize(d)
    LD[i] = annualize(ld)
```

```python
R = pd.DataFrame.from_dict(R, orient='index', columns=['Modest Raise'])
LR = pd.DataFrame.from_dict(LR, orient='index', columns=['Large Raise'])
D = pd.DataFrame.from_dict(D, orient='index', columns=['Modest Drop'])
LD = pd.DataFrame.from_dict(LD, orient='index', columns=['Large Drop'])
df = pd.concat([R, LR, D, LD], axis = 1)
sns.lineplot(data=df, palette="icefire")
plt.title('S&P500 Monthly Lagged Return vs FED Rate Change')
plt.axhline(round(annual['SP500 annual return'].mean(), 4), color = 'red', alpha=0.4, linestyle='--', linewidth=2)
sns.set(rc={'figure.figsize':(8,5)})
plt.xlabel("Lags (month)")
plt.ylabel("Annualized Return")
plt.tight_layout()
plt.savefig('Graph/S&P500 Monthly Lagged Return vs FED.png')
```

On the monthly level, we can track the average lagged return before and after FED movement. 
- The market tend to perform above the average return under a modest raise in FED rate (less than 0.5%). We also see a small dip after the rate hike. 
- The market tend to be volitile when FED decided to have a large raise (greater than 0.5%) and went through several sycle of up and down following the actual rate hike.
- The market tend to perform around the average return under a modest drop in FED rate (less than 0.5%). We also see a small rally going into the FED decision.
- The market tend to react extremely volitle when the FED decided to a large drop (greater than 0.5%). We see the market in a recession going into the rate drop and going on a short rally, from down 30% on average a month before to up 30% two months after the actual rate drop.

### Yield Spread 
Yield curve inversion takes place when the longer term yields falls much faster than short term yields. This is a sign that investors are expecting a slowing economy. They think short term conditions are better than those that are further out.

```python
sns.histplot(data=month, x="T10Y2Y", bins=10)
```

```python
month['Yield Spread level'] = np.where(month['T10Y2Y'] < 0, 'Inverted',
                                np.where(month['T10Y2Y'] < 1, 'low', 
                                     np.where(month['T10Y2Y'] < 2, 'mid', 'high')))
yieldspread_return = month.groupby('Yield Spread level').mean()['SP500 monthly return'].to_frame()
reorderlist = ['Inverted', 'low', 'mid', 'high']
yieldspread_return = yieldspread_return.reindex(reorderlist)
yieldspread_return
```

```python
month['Yield Curve Inversion'] = np.where(month['T10Y2Y'] <= 0, 1, 0)
month.groupby('Yield Curve Inversion').mean()['SP500 monthly return']
```

```python
month['Yield Curve Inversion'].value_counts()
```

```python
df = month['Yield Curve Inversion'][month['Yield Curve Inversion'] == 1].to_frame()
sns.swarmplot(x=df.index, data=df)
plt.xticks(rotation=45)
plt.title('Yield Curve Inversion across time')
```

```python
def delay_return_YieldSpread(lag = 1):
    inversion_delay = []
    normal_delay = []
    for index, row in month.iterrows():
        # print(index)
        if row['Yield Curve Inversion'] == 1:
            inversion_delay.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['Yield Curve Inversion'] == 0:
            normal_delay.append(month['SP500 monthly return'].shift(-lag).loc[index])
    inversion_delay_c = [x for x in inversion_delay if str(x) != 'nan']
    normal_delay_c = [x for x in normal_delay if str(x) != 'nan']
    return round(np.mean(inversion_delay_c), 4), round(np.mean(normal_delay_c), 4)
```

```python
invert = {}
not_invert  = {}
for i in range(-12, 13):
    inversion, no_inversion = delay_return_YieldSpread(i)
    invert[i] = annualize(inversion)
    not_invert[i] = annualize(no_inversion)
```

```python
df = pd.DataFrame.from_dict(invert, orient='index', columns=['Inversion Lagged Return'])
df['Normal Yield Spread return'] = pd.Series(not_invert)
sns.lineplot(data=df)
plt.axhline(np.mean(list(not_invert.values())), color = 'red', alpha=0.4, linestyle='--', linewidth=2)
plt.title('S&P 500 monthly Lagged Return vs Yield Curve Inversion')
sns.set(rc={'figure.figsize':(8,5)})
plt.xlabel("Lags (month)")
plt.ylabel("Annualized Return")
plt.tight_layout()
plt.savefig('Graph/S&P500 Annual Lagged Return vs Inversion.png')
```

We can see from the plot that yield curve inversion signals a potential market downturn. The normal yield curve spread is around the market return. We see the market outperforming around 4 months going into the yield curve inversion as investors anticipate FED policies outlook to the significant impact yield curve. We notice a market rally after the yield curve actually inverted, but the overall market is in a downtrend after the yield curve inversion.     

### Volume
We might see volume correlate with the actual return of the market. We can adjust our porfolio based on the volume signal.

```python
sns.histplot(data=month, x="Volume", bins=20)
```

```python
# monthly return vs volume
month['Volume level'] = np.where(month['Volume'] < 2.5e+09, 'Normal',
                                            np.where(month['Volume'] < 5e+09, 'Medium', 'High'))
volume_return = month.groupby('Volume level').mean()['SP500 monthly return'].to_frame()
reorderlist = ['Normal', 'Medium', 'High']
volume_return = volume_return.reindex(reorderlist)
volume_return
```

```python
def delay_return_volume(lag = 1):
    normal_lag = []
    medium_lag = []
    high_lag = []
    for index, row in month.iterrows():
        # print(index)
        if row['Volume level'] == 'Normal':
            normal_lag.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['Volume level'] == 'Medium':
            medium_lag.append(month['SP500 monthly return'].shift(-lag).loc[index])
        elif row['Volume level'] == 'High':
            high_lag.append(month['SP500 monthly return'].shift(-lag).loc[index])
    normal_lag_c = [x for x in normal_lag if str(x) != 'nan']
    medium_lag_c = [x for x in medium_lag if str(x) != 'nan']
    high_lag_c = [x for x in high_lag if str(x) != 'nan']
    return round(np.mean(normal_lag_c), 4), round(np.mean(medium_lag_c), 4), round(np.mean(high_lag_c), 4) 
```

```python
N = {}
M  = {}
H = {}
for i in range(-12, 13):
    n, m, h = delay_return_volume(i)
    N[i] = annualize(n)
    M[i] = annualize(m)
    H[i] = annualize(h)
```

```python
N = pd.DataFrame.from_dict(N, orient='index', columns=['Normal'])
M = pd.DataFrame.from_dict(M, orient='index', columns=['Medium'])
H = pd.DataFrame.from_dict(H, orient='index', columns=['High'])

df = pd.concat([N, M, H], axis = 1)
sns.lineplot(data=df)
plt.axhline(round(annualize(month['SP500 monthly return'].mean()), 4), color = 'red', alpha=0.4, linestyle='--', linewidth=2)
plt.title('S&P 500 Monthly Lagged Return vs Volume')
sns.set(rc={'figure.figsize':(8,5)})
plt.xlabel("Lags (month)")
plt.ylabel("Annualized Return")
plt.tight_layout()
plt.savefig('Graph/S&P 500 monthly lagged return vs Volume.png')
```

We have some interesting results from the plot above.
- Normal volume has the market return 
- Medium volume outperforms the market before dipping below the market return after the medium volume due to some events
- High volume has the most volitle market. We see the market underperform significantly before going on a rally before the high volume due to some events
