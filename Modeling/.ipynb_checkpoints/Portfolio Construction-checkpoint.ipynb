```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight') # change the style of the plot

# API 
import yfinance as yf
```

```python
sectors = yf.Tickers("XLK, XLV, XLF, XLY, XLC, XLI, XLP, XLE, XLU, XLRE, XLB") 
sectors = sectors.history(period="max")
sectors = sectors['Close']
cols = ['XLK', 'XLV', 'XLF', 'XLY', 'XLC', 'XLI', 'XLP', 'XLE', 'XLU', 'XLRE', 'XLB']
sectors = sectors[cols]
sectors.head()
```

```python
# function that's used to annualize a monthly return 
def annualize(monthly):
    return round((1 + monthly) ** 12 - 1, 4)
```

### Portfolio Optimization
Formulation:

```python
returnPortfolio = sectors.pct_change()
varianceMatrix = returnPortfolio.cov() 
varianceMatrix
```

```python
portfolioReturns = []
portfolioVolitility = []
portfolioWeights = []
numAssets = len(sectors.columns)
numPortfolios = 10000
```

```python
re = sectors.resample('Y').last().pct_change().mean()
```

```python
for p in range(numPortfolios):
    weights = np.random.random(numAssets)
    weights = weights/ np.sum(weights)
    portfolioWeights.append(weights)
    
    returns = np.dot(weights, re)
    portfolioReturns.append(returns)
    
    # var = varianceMatrix.mul(weights, axis = 0).mul(weights, axis=1).sum().sum()
    var = np.dot(weights.T, np.dot(varianceMatrix, weights))
    sd = np.sqrt(var)
    annualizedSD = sd * np.sqrt(250)
    portfolioVolitility.append(annualizedSD)
```

```python
data = {'Return': portfolioReturns, 'Volatility': portfolioVolitility}
for counter, symbol in enumerate(sectors.columns.tolist()):
    data[symbol+' weight'] = [w[counter] for w in portfolioWeights]
```

```python
pV1 = pd.DataFrame(data)
pV1.head()
```

```python
minVolPort = pV1.iloc[pV1['Volatility'].idxmin()]
minVolPort
```

```python
minVolW = pd.DataFrame(minVolPort[2:])
minVolW = minVolW.rename(columns={minVolPort.name: 'Sector Weight'})
minVolW = minVolW.sort_values(by=['Sector Weight'], ascending=False)
minVolW
```

```python
minVolW.loc['Others',:] = minVolW[minVolW['Sector Weight'] <= 0.06].sum()
minVolW = minVolW[minVolW['Sector Weight'] > 0.06]
minVolW
```

```python
labels = minVolW.index
colors = sns.color_palette('flare')
explode = [0.1,.02,0.02,0.02,0.02,0.02,0.02]
plt.pie(minVolW['Sector Weight'], labels=labels, colors = colors, autopct = '%0.0f%%', explode=explode)
plt.tight_layout()
plt.title('Minimum Volatility Equity Allocation')
plt.tight_layout()
plt.savefig('../Graph/Minimum volatility Allocation.png')
plt.show()
```

```python
rf = 0.02
sharpe = pV1.iloc[((pV1['Return'] - rf) / pV1['Volatility']).idxmax()]
sharpe
```

```python
hSharpe = pd.DataFrame(sharpe[2:])
hSharpe = hSharpe.rename(columns={sharpe.name: 'Sector Weight'})
hSharpe = hSharpe.sort_values(by=['Sector Weight'], ascending=False)
hSharpe
```

```python
hSharpe.loc['Others',:] = hSharpe[hSharpe['Sector Weight'] <= 0.06].sum()
hSharpe = hSharpe[hSharpe['Sector Weight'] > 0.06]
hSharpe
```

```python
labels = hSharpe.index
colors = sns.color_palette('flare')
explode = [0.1,.02,0.02,0.02,0.02,0.02,0.02]
plt.pie(hSharpe['Sector Weight'], labels=labels, colors = colors, autopct = '%0.0f%%', explode=explode)
plt.tight_layout()
plt.title('Highest Sharpe Ratio Equity Allocation')
plt.tight_layout()
plt.savefig('../Graph/Highest Sharpe Ratio Allocation.png')
plt.show()
```

We have to have an accurate inflation estimate to know the nominal market return (real return plus inflation). Then, we can have a more accurate sector allocation. Say, the average inflation for 2022 is 7.5%. Then, the expected annual return of the S&P index is going to be 11.74%. Let's find the best portfolio based on this expected return.   

```python
pV1[(pV1['Return'] > 0.112) & (pV1['Return'] < 0.12)]
```

```python
opt = pV1[(pV1['Return'] > 0.112) & (pV1['Return'] < 0.12)].iloc[0]
opt
```

```python
optDF = pd.DataFrame(opt[2:])
optDF = optDF.rename(columns={opt.name: 'Sector Weight'})
optDF = optDF.sort_values(by=['Sector Weight'], ascending=False)
optDF
```

```python
optDF[optDF['Sector Weight'] <= 0.05].sum()
```

```python
optDF.loc['Others',:] = optDF[optDF['Sector Weight'] <= 0.05].sum()
optDF = optDF[optDF['Sector Weight'] > 0.05]
optDF
```

```python
labels = optDF.index
colors = sns.color_palette('flare')
explode = [0.1,.02,0.02,0.02,0.02,0.02,0.02]
plt.pie(optDF['Sector Weight'], labels=labels, colors = colors, autopct = '%0.0f%%', explode=explode)
plt.tight_layout()
plt.title('Optimal Equity Allocation based on Estimated Inflation', fontsize=15)
plt.tight_layout()
plt.savefig('../Graph/Optimal Allocation (Inflation).png')
plt.show()
```
