```python
# data analysis, wrangling and preprocessing
import numpy as np 
import pandas as pd

# data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
plt.style.use('fivethirtyeight') # change the style of the plot

# time series analysis
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm

# preprocessing 
from sklearn.preprocessing import MinMaxScaler

# time series forecasting 
from pmdarima.arima import ndiffs
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pmdarima as pm

# model evaluation 
import math
from sklearn.metrics import mean_squared_error
```

### Import dataset
Import dataset, set the index to be a datetime type and the frequency to be `MS` (the start of each month).

```python
dataset = pd.read_csv('../dataset.csv')
dataset.head()
dataset.index = pd.to_datetime(dataset.index)
dataset.index.freq = 'MS'
```

```python
dataset['SP500 Close Lag 1'] = dataset['SP500 Close'].shift(-1)
dataset.head()
```

### Quarterly decomposition and forecasting 
We expect to have a more accurate prediction with larger timeframe, such as quarterly instead of monthly.

```python
# Test for staionarity
def test_stationarity(timeseries):
    # Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    # Plot rolling statistics:
    plt.plot(timeseries, color='blue', label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    # hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
test_stationarity(dataset.loc[:,['SP500 Close Lag 1']].resample('Q').mean())
```

The time series is not stationary because the p-value is 1. We saw several distinct trend in the quarterly index plot:
- Grew steadily and experienced a short rally in the 2000 dot com bubble
- Recovered to the 2000 level before the financial crisis hit around 2008 
- Had a strong recovery and went tremendous rally from the low of 2008 financial crisis 
- The expansionary monetary policy to recover from the pandemic boost the market on a strong bull run

```python
decomposition = sm.tsa.seasonal_decompose(dataset.loc[:,['SP500 Close']].resample('Q').mean(), extrapolate_trend='freq')
fig = decomposition.plot()
plt.gcf().set_size_inches(10, 8)
```

```python
# Eliminate trend
df_log = np.log(dataset.loc[:,['SP500 Close']].resample('Q').mean())
df_log_lag = np.log(dataset.loc[:,['SP500 Close Lag 1']].resample('Q').mean())
# df_log
moving_avg = df_log.rolling(12).mean()
std_dev = df_log.rolling(12).std()
plt.legend(loc='best')
plt.title('Moving Average')
plt.plot(std_dev, color ="black", label = "Standard Deviation")
plt.plot(moving_avg, color="red", label = "Mean")
plt.legend()
plt.show()
```

```python
# split data into training and testing set
train_data = df_log_lag[:int(len(df_log_lag)*0.9)]
test_data = df_log[int(len(df_log)*0.9):]
plt.figure(figsize=(10,6))
plt.grid(True)
plt.xlabel('Dates')
plt.ylabel('Closing Prices')
plt.plot(df_log_lag, 'green', label='Train data')
plt.plot(test_data, 'blue', label='Test data')
plt.legend()
```

```python
kpss_diffs = ndiffs(train_data, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(train_data, alpha=0.05, test='adf', max_d=6)
n_diffs = max(adf_diffs, kpss_diffs)

print(f"Estimated differencing term: {n_diffs}")
```

```python
model_autoARIMA = pm.auto_arima(train_data, start_p=0, start_q=0,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=4,              # frequency of series
                      d=n_diffs,           # let model determine 'd'
                      seasonal=False,    # With Seasonality
                      start_P=0, 
                      D=None, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
print(model_autoARIMA.summary())
model_autoARIMA.plot_diagnostics(figsize=(15,8))
plt.show()
```

```python
model = model_autoARIMA 
def forecast_one_step():
    fc, conf_int = model.predict(n_periods=1, return_conf_int=True)
    return (
        fc.tolist()[0],
        np.asarray(conf_int).tolist()[0])

forecasts = []
confidence_intervals = []

# print(test_data)
for new_ob in test_data.values:
    fc, conf = forecast_one_step()
    forecasts.append(fc)
    confidence_intervals.append(conf)

    # Updates the existing model with a small number of MLE steps
    model.update(new_ob)
```

```python
print(len(forecasts))
print(test_data.shape)
```

```python
# Make as pandas series
fc_series = pd.Series(forecasts, index=test_data.index)
lower_series = pd.Series(np.exp([i[0] for i in confidence_intervals]), index=test_data.index)
upper_series = pd.Series(np.exp([i[1] for i in confidence_intervals]), index=test_data.index)
# Plot
plt.figure(figsize=(10,5))
plt.plot(np.exp(train_data), label = 'training data')
plt.plot(np.exp(test_data), color = 'blue', label='Actual Index Level')
plt.plot(np.exp(fc_series), color = 'orange',label='Predicted Index Level')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.10)
plt.title('S&P 500 Index ARIMA Prediction')
plt.xlabel('Time')
plt.ylabel('S&P Index')
plt.legend(loc='upper left', fontsize=12)
plt.tight_layout()
plt.savefig('../Graph/S&P500 ARIMA Prediction(Q).png')
plt.show()
```

```python
mse = mean_squared_error(np.exp(test_data), np.exp(fc_series))
rmse = math.sqrt(mse)
print(rmse)
plt.plot(np.exp(test_data), color = 'blue', label='Actual Index Level')
plt.plot(np.exp(fc_series), color = 'orange',label='Predicted Index Level')
plt.xticks(rotation = 45) 
```

### SARIMAX model

```python
print(dataset.shape)
dataset.head()
```

```python
dataset_for_prediction = dataset.copy()
dataset_for_prediction.head(3)
```

```python
dataset_for_prediction = dataset_for_prediction.dropna()
```

```python
sc_in = MinMaxScaler(feature_range=(0, 1))
scaled_input = sc_in.fit_transform(dataset_for_prediction[['CPIAUCSL_PC1', 'UNRATE', 'FEDFUNDS', 'T10Y2Y', 'High', 'Low',
                                                            'Volume', 'ADX', 'RSI', 'BB width']])
scaled_input = pd.DataFrame(scaled_input, index = dataset_for_prediction.index, columns = {'CPIAUCSL_PC1', 'UNRATE', 
                                                'FEDFUNDS', 'T10Y2Y', 'High', 'Low', 'Volume', 'ADX', 'RSI', 'BB width'})
X = scaled_input
```

```python
print(X.shape)
X.head()
```

```python
sc_out = MinMaxScaler(feature_range=(0, 1))
scaler_output = sc_out.fit_transform(dataset_for_prediction[['SP500 Close Lag 1']])
scaler_output = pd.DataFrame(scaler_output, dataset_for_prediction.index, columns = {'SP500 Close Lag 1'})
y = scaler_output
```

```python
train_size = int(len(dataset) * 0.9)
test_size = int(len(dataset)) - train_size
train_X, train_y = X[:train_size].dropna(), y[:train_size].dropna()
test_X, test_y = X[train_size:].dropna(), y[train_size:].dropna()
test_size = test_y.shape[0]
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

```python
train_index = train_y.index
test_index = test_y.index
```

```python
def test_adf(series, title=''):
    dfout = {}
    dftest = sm.tsa.adfuller(series.dropna(), autolag = 'AIC', regression = 'ct')
    for key,val in dftest[4].items():
        dfout[f'critical value ({key})'] = val
    if dftest[1] <= 0.05:
        print("Strong evidence against Null Hypothesis")
        print("Reject Null Hypothesis - Data is Stationary")
        print("Data is Stationary", title)
    else:
        print("Strong evidence for  Null Hypothesis")
        print("Accept Null Hypothesis - Data is not Stationary")
        print("Data is NOT Stationary for", title)
```

```python
test_adf(test_y, " Stock Price")
```

```python
test_adf(test_y.diff(), " Stock Price")
```

```python
autoSARIMAX = pm.auto_arima(train_y, exogenous = train_X, m=12, seasonal=True, d=1, trace=True, error_action='ignore', 
                        suppress_warnings=True, stepwise=True)
autoSARIMAX.summary()
```

```python
model = SARIMAX(train_y, exog=train_X, order=(0,1,0), seasonal_order=(0,0,0,12), enforce_invertibility=False, 
                enforce_stationarity=False)
results = model.fit()
```

```python
forecast = results.forecast(steps=test_size, exog=test_X)
# forecast
forecast_ci = results.get_forecast(steps=test_size, exog=test_X)
# forecast_ci
ci = forecast_ci.conf_int(alpha=0.05)
forecast = sc_out.inverse_transform(np.array(forecast).reshape(-1, 1))
ci = sc_out.inverse_transform(np.array(ci).reshape(-1, 2))
forecast = np.reshape(forecast, (1, -1)).tolist()[0]
# forecast
```

```python
test_y = sc_out.inverse_transform(np.array(test_y).reshape(1, -1)).tolist()[0]
train_y = sc_out.inverse_transform(np.array(train_y).reshape(1, -1)).tolist()[0]
```

```python
ci_low = [i[0] for i in ci]
ci_high = [i[1] for i in ci]
```

```python
d = {'y_test': test_y, 'Forecast': forecast, 'Conf Low': ci_low, 'Conf High': ci_high}
df = pd.DataFrame(d, index=test_index)
# df.head()
ty = {'y_train':train_y}
train_y_df = pd.DataFrame(ty, index=train_index)
# train_y_df.head()
```

```python
# Make as pandas series
fc_series = pd.Series(df['Forecast'], index=df.index)
lower_series = pd.Series(df['Conf Low'], index=df.index)
upper_series = pd.Series(df['Conf High'], index=df.index)

# Plot
plt.figure(figsize=(10,5))
plt.plot(train_y_df, label='training data')
plt.plot(df['y_test'], color = 'blue', label='Actual Index Level')
plt.plot(fc_series, color = 'orange',label='Predicted Index Level')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.10)
plt.title('S&P 500 Index SARIMAX Prediction')
plt.xlabel('Time')
plt.ylabel('S&P Index')
plt.legend(loc='upper left', fontsize=12)
plt.tight_layout()
plt.savefig('../Graph/S&P500 SARIMAX Prediction(M).png')
plt.show()
```

```python
mse = mean_squared_error(test_y, forecast)
rmse = math.sqrt(mse)
print(rmse)
plt.plot(df['y_test'], color = 'blue', label='Actual Index Level')
plt.plot(df['Forecast'], color = 'orange',label='Predicted Index Level')

fc_series = pd.Series(df['Forecast'], index=df.index)
lower_series = pd.Series(df['Conf Low'], index=df.index)
upper_series = pd.Series(df['Conf High'], index=df.index)
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.10)
plt.xticks(rotation = 45) 
```
